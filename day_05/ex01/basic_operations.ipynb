```python
import pandas as pd
```

### create a dataframe views

```python
views = pd.read_csv("../data/feed-views.log", sep="\t", header=None, names=["datetime", "user"])
views
```

```python
views.info()
```

```python
views["datetime"] = pd.to_datetime(views["datetime"])
views["year"] = views.datetime.dt.year
views["month"] = views.datetime.dt.month
views["day"] = views.datetime.dt.day
views["hour"] = views.datetime.dt.hour
views["minute"] = views.datetime.dt.minute
views["second"] = views.datetime.dt.second
```

```python
views.info()
```

```python
views.head()
```

### create the new column daytime

```python
intervals = [0, 4, 7, 11, 17, 20, 24]
labels = [ "night", "early morning", "morning", "afternoon", "early evening", "evening"]
views["daytime"] = pd.cut(views["hour"], bins=intervals, labels=labels, 
                       right=False, include_lowest=True)
views
```

```python
views.set_index('user', inplace=True)
views
```

```python
views.count()
```

```python
views["daytime"].value_counts()
```

### sort values in your dataframe

```python
views.sort_values(["hour","minute","second"], inplace=True)
views
```

### calculate the minimum and maximum for the hours and the mode for the daytime categories

```python
views[views["daytime"]=="night"]["hour"].max()
```

```python
views[views["daytime"]=="morning"]["hour"].min()
```

```python
views[views["hour"] == views[views["daytime"]=="night"]["hour"].max()]
```

```python
views[views["hour"] == views[views["daytime"]=="night"]["hour"].max()].index.values[0]
```

```python
views[views["hour"] == views[views["daytime"]=="morning"]["hour"].min()].index.values[0]
```

```python
views["hour"].mode()
```

```python
views["daytime"].mode()
```

### show the 3 earliest hours in the morning and the corresponding usernames and the 3 latest hours and the usernames using nsmallest() and nlargest()

```python
views[views["daytime"]=="morning"].nsmallest(3, 'hour')["hour"]
```

```python
views[views["daytime"]=="morning"].nlargest(3, 'hour')["hour"]
```
